```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind 
```

```python
df = pd.read_csv('C:/Users/MazaherGhorbani/Desktop/Interview Cases/Trivago/case_study_marketing_20210201.csv')
dimdate = pd.read_excel('C:/Users/MazaherGhorbani/Desktop/Interview Cases/Trivago/dimdates.xlsx')
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df_joined = df.merge(dimdate, how='left', left_on='ymd', right_on='DateNum')
df_joined.head(3)
```

```python
df_joined.columns
```

```python
df_september = df_joined[df_joined['MonthName']=='September']
df_october = df_joined[df_joined['MonthName']=='October']
df1 = df_october['bookings'].dropna()
df2 = df_september['bookings'].dropna()
```

```python
ttest_ind(df1,df2)
```

```python
df3 = df_october['impressions'].dropna()
df4 = df_september['impressions'].dropna()
```

```python
ttest_ind(df3,df4)
```
